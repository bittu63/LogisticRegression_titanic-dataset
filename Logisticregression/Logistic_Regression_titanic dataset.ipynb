```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd

train_data = pd.read_csv("C:/Users/biswa/OneDrive/Desktop/0000000000002429_training_titanic_x_y_train.csv")
test_data = pd.read_csv("C:/Users/biswa/OneDrive/Desktop/0000000000002429_test_titanic_x_test.csv")
```

```python
train_data['Name']
title_in_train = []

for i in train_data['Name']:
    title_in_train.append(i.split()[1][0:-1])
    
title_in_train = np.array(title_in_train)

title_in_train[title_in_train == "Mr"]=0
title_in_train[title_in_train == "Mrs"]=1
title_in_train[title_in_train == "Miss"]=2
title_in_train[title_in_train == "Master"]=3
title_in_train[(title_in_train != '0')&(title_in_train != '1')&(title_in_train != '2')&(title_in_train != '3')]=4
title_in_train = np.array(title_in_train,dtype =int)
title_in_train
```

```python
test_data['Name']
title_in_test = []

for i in test_data['Name']:
    title_in_test.append(i.split()[1][0:-1])
    
title_in_test = np.array(title_in_test)

title_in_test[title_in_test == "Mr"]=0
title_in_test[title_in_test == "Mrs"]=1
title_in_test[title_in_test == "Miss"]=2
title_in_test[title_in_test == "Master"]=3
title_in_test[(title_in_test != '0')&(title_in_test != '1')&(title_in_test != '2')&(title_in_test != '3')]=4
title_in_test = np.array(title_in_test,dtype =int)
title_in_test
```

```python
train_data.loc[train_data['Sex']=='female', 'Sex']=0
train_data.loc[train_data['Sex']=='male', 'Sex']=1
train_data.loc[train_data['Embarked'].isnull(), 'Embarked']='S'
train_data.loc[train_data['Embarked']=='S', 'Embarked']=1
train_data.loc[train_data['Embarked']=='C', 'Embarked']=2
train_data.loc[train_data['Embarked']=='Q', 'Embarked']=3
train_data.loc[train_data['Cabin'].isnull(), 'Cabin']=0
train_data.loc[train_data['Cabin']!=0, 'Cabin']=1
train_data.loc[train_data['Age'].isnull(), 'Age']= train_data.loc[train_data['Age'].notnull(), 'Age'].median()
train_data=train_data.drop('Ticket', axis=1)
train_data= train_data.drop('Name', axis=1)

training_data= train_data.values
print(train_data.values)
x = training_data[:, :-1]
y = training_data[:, -1]
y = y.astype('int')
x = np.append(x, title_in_train.reshape(-1, 1), axis=1)
x.shape
```

```python
test_data.loc[test_data['Sex']=='female', 'Sex']=0
test_data.loc[test_data['Sex']=='male', 'Sex']=1
test_data.loc[test_data['Embarked'].isnull(), 'Embarked']='S'
test_data.loc[test_data['Embarked']=='S', 'Embarked']=1
test_data.loc[test_data['Embarked']=='C', 'Embarked']=2
test_data.loc[test_data['Embarked']=='Q', 'Embarked']=3
test_data.loc[test_data['Cabin'].isnull(), 'Cabin']=0
test_data.loc[test_data['Cabin']!=0, 'Cabin']=1
test_data.loc[test_data['Age'].isnull(), 'Age']=test_data.loc[train_data['Age'].notnull(),'Age'].median()
test_data=test_data.drop('Ticket', axis=1)
test_data= test_data.drop('Name', axis=1)

testing_data=test_data.values
testing_data=np.append(testing_data, title_in_test.reshape(-1, 1), axis=1)
testing_data.shape
```

```python
clf=LogisticRegression(solver = 'lbfgs',max_iter=1000000, tol=0.00001)
clf.fit(x, y)
```

```python
predictions =clf.predict(testing_data)
```

```python
clf.score(x,y)
```
